```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive
drive.mount('/content/drive') # mount personal G-Drive

# /content/drive/MyDrive/UMKC_Fa24/datasets/[filename] <- to import dataset from Google Drive
```

```python
df_country = pd.read_csv('/content/drive/MyDrive/UMKC_Fa24/datasets/country_complete.csv')
df_country.head(5)
```

# Visualize one feature

## one categorical feature

```python
# bar chart
sns.histplot(x='Continent', data=df_country, shrink=0.7);
```

```python
# bar chart
sns.histplot(x='Internet access', data=df_country, shrink=0.7);
```

## one numerical feature

```python
# bar chart/plot
sns.histplot(x='Fertility', data=df_country, shrink=.8, color='red', stat='percent');
```

```python
sns.boxplot(x='Fertility', data=df_country, color='yellow')
```

```python
sns.kdeplot(x='Fertility', data=df_country, color='green')
```

```python
sns.histplot(x='Fertility', data=df_country, stat='density', kde=True, color='orange')
```

# Visualize two features

## two numerical features

```python
sns.scatterplot(x='Internet', y='Years', data=df_country);
```

## two categorical features

```python
# group bar chart
sns.countplot(x='Continent', data=df_country, hue='Internet access');
```
